```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos (reemplaza esto con la carga de tus datos)
url = 'https://raw.githubusercontent.com/LabSWPP12023S2G2/TPInicial/main/datasetUNC.csv'
data = pd.read_csv(url, delimiter=';')
columns_to_drop = ['SUB PERIODS']
data = data.drop(columns=columns_to_drop)
data = data.dropna(axis=0)
data.drop(data[data['PROVINCE'] == 'Otro'].index, inplace=True)
data.drop(data[data['PROVINCE'] == 'other'].index, inplace=True)
```

```python
# Aplicar K-Means para clasificar las provincias en grupos
province_data = data.groupby('PROVINCE').mean()
kmeans = KMeans(n_clusters=5, random_state=0).fit(province_data[['SUIC RISK']])

# Asignar un color a cada provincia en base a los clusters
province_data['Cluster'] = kmeans.labels_
```

```python
province_data
```

```python
# Definir colores de los clusters
cluster_colors = {0: 'red', 1: 'blue', 2: 'green', 3: 'purple', 4: 'orange'}

# Visualización de las provincias por cluster
tsne = TSNE(n_components=2, random_state=0, perplexity=10)
tsne_data = tsne.fit_transform(province_data.drop(columns=['Cluster']))
province_data['tsne_x'] = tsne_data[:, 0]
province_data['tsne_y'] = tsne_data[:, 1]

# Gráfico de dispersión
plt.figure(figsize=(12, 8))
for cluster, color in cluster_colors.items():
    cluster_data = province_data[province_data['Cluster'] == cluster]
    plt.scatter(cluster_data['tsne_x'], cluster_data['tsne_y'],
                label=f'Cluster {cluster}', color=color)

# Etiquetas de provincias por nombre
for index, row in province_data.iterrows():
    province_name = index
    plt.text(row['tsne_x'], row['tsne_y'], province_name, fontsize=10, color=cluster_colors[row['Cluster']])

plt.title('Clustering de Provincias')
plt.xlabel('t-SNE Dimensión 1')
plt.ylabel('t-SNE Dimensión 2')
plt.legend()
plt.show()
```

```python
# Definir la función para asignar rangos de riesgo
def map_risk_range(value):
    if value <= 15:
        return 'Riesgo Bajo'
    elif value <= 30:
        return 'Riesgo Medio-Bajo'
    elif value <= 50:
        return 'Riesgo Mediano'
    elif value <= 65:
        return 'Riesgo Alto'
    else:
        return 'Riesgo Muy Alto'

# Aplicar la función para crear la columna "Rango Riesgo"
data['Rango Riesgo'] = data['SUIC RISK'].apply(map_risk_range)
```

```python
# Encontrar el índice de la fila con el máximo riesgo
indice_max_riesgo = data['SUIC RISK'].idxmax()

# Acceder a la fila completa con el máximo riesgo
fila_max_riesgo = data.loc[indice_max_riesgo]

print("Fila con máximo riesgo:")
print(fila_max_riesgo)
```

```python
# Crear un mapeo de colores y tamaños basado en la cantidad de casos
color_palette = {'Riesgo Bajo': 'lightgreen', 'Riesgo Medio-Bajo': 'gold', 'Riesgo Mediano': 'orange', 'Riesgo Alto': 'red', 'Riesgo Muy Alto': 'darkred'}
size_scale = {'Riesgo Bajo': 20, 'Riesgo Medio-Bajo': 40, 'Riesgo Mediano': 70, 'Riesgo Alto': 100, 'Riesgo Muy Alto': 150}

# Orden correcto de valores para la leyenda
hue_order = ['Riesgo Bajo', 'Riesgo Medio-Bajo', 'Riesgo Mediano', 'Riesgo Alto', 'Riesgo Muy Alto']

# Crear la visualización con los mapeos personalizados y orden de leyenda
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='SUIC RISK', y='PROVINCE', hue='Rango Riesgo', palette=color_palette, size='Rango Riesgo', sizes=size_scale, legend='brief', hue_order=hue_order)
plt.xlabel('Riesgo')
plt.ylabel('Provincia')
plt.title('Riesgo por Provincia')
plt.xticks(rotation=0)
plt.tight_layout()
plt.legend(title='Rango de Riesgo', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
